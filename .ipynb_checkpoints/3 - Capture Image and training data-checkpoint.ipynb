```python
import cv2
import numpy as np
```

```python
# reading video stream
import cv2

cap = cv2.VideoCapture(0)  ## (0) is id ( 0-defalut webcam)
face_cascade = cv2.CascadeClassifier("/home/ankur/Documents/haarcascade_frontalface_default.xml")

skip = 0

face_data = []

dataset_path = '/home/ankur/Desktop/Face Data/'

file_name = input('Enter name of person: ')

while True :
    ret,frame = cap.read()
    gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    if ret==False :
        continue
        
    faces = face_cascade.detectMultiScale(gray_frame,1.3,5) ##(scalingFactor,No of Neighbours)
    #print(faces)  ## to check np array getting stored
    
    
    ## Now we have to store the largest face .We will sort by area ,
    ## area = w * h 
    ## introducing lambda function here
    
    faces = sorted(faces,key =lambda f:f[2]*f[3])
    
    ##picking last face(that would be lof argest area)
    
    for (x,y,w,h) in faces[-1:] :
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        
    
      
    
    
    ## extracting face (crop)
    offset = 10   ## providing 10 pixel of margin to all four sides
    face_section = frame[y-offset:y+h+offset,x-offset:x+w+offset]
    face_sectione = cv2.resize(face_section,(100,100))
    
    
    
    
    
    
    ## store every 10th face  (10th frame)
    skip+=1
    if(skip%10==0):
        face_data.append(face_section)
        print(len(face_data))
        
    cv2.imshow('Frame',frame)
    cv2.imshow("face Section",face_section)
    
    
    key_pressed = cv2.waitKey(1) & 0xFF   ## converting 32 byte into  8 bytes and then comparing with ascii value
    
    ## wait for user input (q) , then you will stop loop
    
    if key_pressed == ord('q'):
        break

## convert face list into numpy array

face_data = np.asarray(face_data)
face_data = face_datae.reshape((face_data.shape[0],-1))

## saving array

np.save(dataset_path+file_name+'.npy',face_data)
        
 
print('Data saved at "+dataset_path+file_name+.npy"')

cap.release()
cv2.destroyAllWindows()



## scaleFactor = parameter specifying how much image size is reduced at each image scale
## used to create scale pyramid

## neighbours - specifying how many neighbours each candidate rectange should have to retain it
## more value - less detection but high quality (keep 3-6)
```
